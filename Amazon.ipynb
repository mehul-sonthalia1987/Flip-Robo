```python
from bs4 import BeautifulSoup
import requests
import re
HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
url = 'https://www.amazon.in/s?k=mobiles+under+20000&ref=nb_sb_noss_2'
response = requests.get(url,headers=HEADERS)
soup = BeautifulSoup(response.text, 'lxml')
Product_Name=[]
Product_Price=[]
Product_ImageURL=[]
Product_Rating=[]
Name = soup.select('span.a-size-medium.a-color-base.a-text-normal')
for i in Name:
  Product_Name.append(i.text.replace('\n', ''))

Price=soup.select('span.a-price-whole')
for i in Price:
  Product_Price.append(i.text.replace('\n', ''))
Product_Price

Product_ImageURL = soup.select('div.a-section.aok-relative.s-image-fixed-height')

import pandas as pd
Amazon = pd.DataFrame({})
Amazon['Name']=Product_Name
Amazon['Price']=Product_Price
Amazon['URL']=Product_ImageURL
Amazon
```

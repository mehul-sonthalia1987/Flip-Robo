```python
from bs4 import BeautifulSoup
import requests
import re
url = 'https://internshala.com/fresher-jobs'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

job_title=[]
Name_Temp=[]
company_name=[]
Salary_Temp=[]
CTC=[]
CTC_Temp2=[]
CTC_Temp=[]
apply_date=[]

Title = soup.select('div.heading_4_5.profile')
for i in Title:
  job_title.append(i.text.replace('\n', ''))

Company = soup.select('div.heading_6.company_name')
for i in Company:
  Name_Temp.append(i.text.replace('\n', ''))
for i in Name_Temp:
  company_name.append(i.replace(' ','')) 

Salary = soup.select('div.other_detail_item')
for i in Salary:
  Salary_Temp.append(i.text.replace('\n', ''))

for i in Salary_Temp[1:120:3]:
   CTC_Temp.append(i.replace('CTC',''))
for i in CTC_Temp:
  CTC_Temp2.append(i.replace('LPA',''))
for i in CTC_Temp2:
  CTC.append(i.replace(' ',''))

for i in Salary_Temp[2:120:3]:
  apply_date.append(i.replace('Apply By',''))


import pandas as pd
Internshala = pd.DataFrame({})
Internshala['Job Title']=job_title
Internshala['Company Name']=company_name
Internshala['CTC(LPA) ']=CTC
Internshala['Apply Date']=apply_date
Internshala
```
